```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv(
    r"C:\Users\HP\Projects\Data-Analyst-Portfolio\Sales-Analysis-Project\data\superstore.csv.csv"
)

df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.columns
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```

```python
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Order Month Name"] = df["Order Date"].dt.month_name()
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
category_sales
```

```python
category_sales.plot(kind="bar", title="Sales by Category", figsize=(8,5))
plt.show()
```

```python
df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Sub-Category")["Sales"].sum().plot(
    kind="bar",
    figsize=(10,6),
    title="Sales by Sub-Category"
)
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(
    kind="bar",
    title="Sales by Region",
    figsize=(6,4)
)
plt.show()
```

```python
df.to_csv(
    r"C:\Users\HP\Projects\Data-Analyst-Portfolio\Sales-Analysis-Project\data\superstore_cleaned.csv",
    index=False
)
```

```python
df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Segment")["Sales"].sum().plot(
    kind="bar",
    figsize=(6,4),
    title="Sales by Customer Segment"
)
plt.ylabel("Total Sales")
plt.show()
```

```python
top_customers = (
    df.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_customers
```

```python
top_customers.plot(
    kind="bar",
    figsize=(10,5),
    title="Top 10 Customers by Sales"
)
plt.show()
```

```python
df["Segment"].value_counts()
```

```python
order_sales = df.groupby("Order ID")["Sales"].sum()
order_sales.describe()
```

```python
pivot_segment_region = pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Segment",
    aggfunc="sum"
)

pivot_segment_region
```

```python
pivot_segment_region.plot(
    kind="bar",
    figsize=(10,6),
    title="Segment-wise Sales by Region"
)
plt.show()
```
